```python
import numpy as np
import pandas as pd
```

```python
training_df = pd.read_csv('/kaggle/input/titanic/train.csv')
display(training_df)
```

```python
testing_df = pd.read_csv('/kaggle/input/titanic/test.csv')
testing_df
```

```python
def clean_data(df):
    df['Age'].fillna(df['Age'].median(), inplace=True)
    df.drop(['Cabin', 'Ticket', 'Name', 'PassengerId'], axis=1, inplace=True)
    return df
```

```python
from sklearn.preprocessing import LabelEncoder
def encode_data(df):
    encoder = LabelEncoder()
    
    cols = ['Sex', 'Embarked']
    
    for col in cols:
        df[col] = encoder.fit_transform(df[col])
    
    return df
    
```

```python
training_df = clean_data(training_df)
training_df
```

```python
training_df = encode_data(training_df)
training_df
```

```python
training_df.isna().sum()
```

```python
X = training_df.drop('Survived', axis=1)
y = training_df['Survived']
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(X, y)
```

```python
testing_df = clean_data(testing_df)
testing_df
```

```python
testing_df = encode_data(testing_df)
```

```python
testing_df
```

```python
testing_df.isna().sum()
```

```python
testing_df['Fare'].fillna(testing_df['Fare'].median(), inplace=True)
```

```python
testing_df.isna().sum()
```

```python
predictions = logreg.predict(testing_df)
```

```python
predictions_dict = {
    "PassengerId": testing_df['PassengerId'],
    "Survived" : predictions
}

predictions_df = pd.DataFrame(predictions_dict)
predictions_df
```

```python
predictions_df.to_csv('/kaggle/working/predictions.csv', index=False)
```
